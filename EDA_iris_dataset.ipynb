```python
import pandas as pd # data processing

import numpy as np # linear algebra

#Visualizations
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = pd.read_csv("Iris.csv")

iris.head()
```

```python
iris['Species'].value_counts()
```

```python
iris.describe()
```

```python
# Calculating the mean and median of each specie by measurement

# Selecting only the relevant columns (excluding 'Id')
iris_selected = iris[['Species', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

# Group by 'Species' and calculate mean and median
iris_selected.groupby('Species').agg(['mean', 'median'])
```

```python
from scipy import stats

# Calculate z-scores
z_scores = stats.zscore(iris['SepalLengthCm'])

# Define threshold for identifying outliers
threshold = 2

# Find outliers
outliers = iris['SepalLengthCm'][np.abs(z_scores) > threshold]

print("Outliers detected in SepalLengthCm:")
print(outliers)
```
